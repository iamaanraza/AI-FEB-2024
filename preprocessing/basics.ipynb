# Preprocessing steps:
1. Imputating missing values (filling out missing data) (`sklearn.impute`)



```python
import pandas as pd 
import numpy as np

a = pd.DataFrame({
    'A': [1, 2, 3, None, None, 23, 56, 67, 78],
    'B':[54, None, 6, None, 1,2,2,5,5],
    'C': ['A','B', np.nan, np.nan, 'B','B','A','A','B'],
    'D':['Purchased','Not Purchased','Purchased','Purchased','Not Purchased','Purchased',
         'Not Purchaesd','Not Purchased','Purchased']
})
a
```

Rule : never impute the missing values in target variables , Drop the rows with missing target variables values.

```python
from sklearn.impute import SimpleImputer 

# object creation
si = SimpleImputer()
# fit the imputer
a[['A','B']] = si.fit_transform(a[['A','B']])
si2 =SimpleImputer(strategy='most_frequent')
a[['C']]= si2.fit_transform(a[['C']])
a
```

2. Encoding Categorical Variable (`sklearn.preprocessing`)
    - Ordinal encoding (`OrdinalEncoder`)
    - Label encoding (`LabelEncoder`) for target variable only
    - One - Hot Encoding (`OneHotEncoder`)

```python
from sklearn.preprocessing import OrdinalEncoder
# object set
oe = OrdinalEncoder()
# tranform kar raha hai and rakh raha hai .
a[['C']] = oe.fit_transform(a[['C']])
a
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
a['D']=le.fit_transform(a[['D']])
a
```

```python
from sklearn.preprocessing import OneHotEncoder

df = pd.DataFrame({
    'City': ['Delhi','Mumbai','Hyderabad','Mumbai','Delhi'],
    'pop': [12,13,14,15,16]
})
he = OneHotEncoder(drop='first', sparse_output=False)
enc_city = he.fit_transform(df[['City']])
df =pd.concat([df, pd.DataFrame(enc_city)], axis=1)
df.drop('City',axis=1, inplace=True)
df
```

```python
he.inverse_transform([[1,0],[0,0]])
```

```python
from sklearn.preprocessing import StandardScaler
df = pd.DataFrame({
    'Salary':[123000,10000,223900,421424],
    'age': [44,34,54,56],
})
sc= StandardScaler()
df[['Salary','age']]= sc.fit_transform(df)
df
```
