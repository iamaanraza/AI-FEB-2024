Simple Linear Regression 

In this regression, there are two variables, i.e., one independent variable and one dependent variable. The independent variable is the input, and the dependent variable is the output.

The equation of the simple linear regression is: y = mx + c

This algorithms tries to find the values of m and c such that the error is minimized.

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/digipodium/Datasets/main/regression/Salary_Data.csv'
df = pd.read_csv(url)
df # head() se bada ans nhi dikhayega
```

```python
df.plot(x='YearsExperience', y='Salary', style='o')
plt.show()
```

no preprocessing is required in this example

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X = df[['YearsExperience']] # 2D martrix (dataframe)
y = df['Salary'] # 1D array (series)

X_train , X_test , y_train, y_test = train_test_split(X,y, test_size=.2)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, median_absolute_error, r2_score

y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
mae = median_absolute_error(y_test, y_pred)
acc = r2_score(y_test, y_pred)

print(f'MSE: {mse:.1f}')
print(f'MAE: {mae:.1f}')
print(f'Accuracy: {acc*100:.1f}')
```

visualize the result 

```python
plt.title('Linear Regression Test Result')
plt.scatter(X_test, y_test, color='gray', label='Actual')
plt.plot(X_test, y_pred, color='red', label='Predicted')
plt.legend()
plt.text(5, 110000, f'Accuracy: {acc*100:.1f}%', bbox={'edgecolor':'red'})
plt.show()
```
